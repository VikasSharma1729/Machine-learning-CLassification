```python
def grades (cols):
    ca,mte,etp_theory,etp_practical,courseattendace = cols[0],cols[1],cols[2],cols[3],cols[4]
    ca = (ca/100)*25
    courseattendace = (courseattendace/100)*25
    total = ((ca+mte+etp_theory+etp_practical+etp_theory+courseattendace)/200)*100
    return total
```

```python
dataset['total_marks'] = dataset[['CA_100','MTT_50','ETT_100','ETP_100','Course_Att']].apply(grades,axis = 1)
```

```python
grade =[0]*dataset.shape[0]
```

```python
dataset["grades"] = grade
```

```python
m = max(dataset['total_marks'])
grade = []
```

```python
#Classifying the datat into different categories.
def marks (cols):
    mark = (cols[0]/m)*100
    
    if((mark>95)):
        return 1;
    elif(mark>90):
        return 2
    elif(mark>83):
        return 3
    elif(mark>75):
        return 4
    elif(mark>60):
        return 5
    elif(mark>54):
        return 8
    elif(mark>33):
        return 9
    else:
        return 10
    
```

```python
dataset['grade'] = dataset[['total_marks']].apply(marks,axis = 1)
```
